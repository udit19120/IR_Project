```python
def extract_data(data, path_to_save):

  user_ID_list = []
  product_ID_list = []
  overall_rating_list = []
  review_list = []

  for sample in data:
    # print(sample.keys())

    if('reviewText' not in sample):
      continue
    userID = sample['reviewerID']
    productID = sample['asin']
    rating = sample['overall']
    review = sample['reviewText']

    user_ID_list.append(userID)
    product_ID_list.append(productID)
    overall_rating_list.append(rating)
    review_list.append(review)
  
  df = pd.DataFrame(list(zip(user_ID_list, product_ID_list, review_list, overall_rating_list)), columns =['UserID', 'ProductID', 'Review', 'Overall Rating'])
  df.to_csv(path_to_save, index=False)
  return df
```

```python
def create_user_item_map(data, path_to_save):

  user_ID_list = []
  product_ID_list = []
  overall_rating_list = []

  for sample in data:
    if('reviewText' not in sample):
      continue
    
    userID = sample['reviewerID']
    productID = sample['asin']
    rating = sample['overall']

    user_ID_list.append(userID)
    product_ID_list.append(productID)
    overall_rating_list.append(rating)

  df = pd.DataFrame(list(zip(user_ID_list, product_ID_list, overall_rating_list)), columns =['UserID', 'ProductID', 'Overall Rating'])
  df.to_csv(path_to_save, index=False)
  return df
```
